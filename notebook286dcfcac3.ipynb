```python
#Percentage of orders more than 1000
np.sum(df['order_amount'] > 1000)/5000*100
```

" Order amount" has a huge amount of standard deviation and 75% of the orders have values smaller than 390.Only 71 orders from 5000(less than 2%)  orders have value more than 1000.  So, the mean value(3145.128) is not a good representative of AOV. The Median is a better metric to describe the order amount status.

```python
#Calculate Median
df['order_amount'].median()
```

Mode is also in a range of around 200.



```python
#Calculate Mode
df['order_amount'].mode()
```

Summary: Only one statical metric like MEAN is not enough to give a great overall view of data patterns. It is better to calculate Median and Mode of data and select one that is the best representative of majority of Order values.  I believe Median is the best metric for the data .
